```python
import os
class Product:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity


class InventorySystem:
    def __init__(self):
        self.inventory = {}

    # View Inventory
    def view_inventory(self):
        if not self.inventory:
            print("\nNo products available.")
        else:
            print("\n----- INVENTORY DASHBOARD -----")
            for product in self.inventory.values():
                print(f"{product.name} : {product.quantity}")

    # Add / Restock Product
    def add_product(self):
        name = input("Enter product name: ")
        qty = int(input("Enter quantity: "))

        if name in self.inventory:
            self.inventory[name].quantity += qty
            print("Stock Updated Successfully!")
        else:
            self.inventory[name] = Product(name, qty)
            print("Product Added Successfully!")

    # Process Sale
    def process_sale(self):
        name = input("Enter product name sold: ")

        if name not in self.inventory:
            print("Product not found!")
            return

        qty = int(input("Enter quantity sold: "))

        if qty <= self.inventory[name].quantity:
            self.inventory[name].quantity -= qty
            print("Sale Processed Successfully!")
        else:
            print("Not enough stock available!")

    # Save Stock Report
    def save_stock_report(self):
        with open("stock.txt", "a") as file:
            file.write("CURRENT STOCK REPORT\n")
            file.write("---------------------\n")

            for product in self.inventory.values():
                file.write(
                    f"{product.name} : {product.quantity}\n"
                )

        print("Stock report saved in stock.txt")


# Main Program
system = InventorySystem()

while True:
    print("\n===== SUPPLY CHAIN & INVENTORY MANAGEMENT =====")
    print("1. View Inventory Dashboard")
    print("2. Inbound Logistics: Restock/Add Product")
    print("3. Outbound Logistics: Process Order Sale")
    print("4. Save Stock Report")
    print("5. Exit System")

    choice = input("Enter your choice: ")

    if choice == "1":
        system.view_inventory()

    elif choice == "2":
        system.add_product()

    elif choice == "3":
        system.process_sale()

    elif choice == "4":
        system.save_stock_report()

    elif choice == "5":
        print("Exiting System...")
        break

    else:
        print("Invalid Choice!")
```
